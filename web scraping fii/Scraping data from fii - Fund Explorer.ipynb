### Let's collect data from Funds Explorer site from our FIIs dataset

```python

from bs4 import BeautifulSoup
import csv
import requests
import pandas as pd
 
page = ""
url = 'https://www.fundsexplorer.com.br/funds/'
```

```python
# This function is going to be important when we'll change among the pages
def get_soup(page):
    response = requests.get(url + page)
    content = response.content
    soup = BeautifulSoup(content,'lxml')
    
    return soup
```

```python
fii_list = pd.read_csv('list_name_fii.csv')
```

```python
fii_list.head(7)
# Let's check the dataset that we scraped before...
```

### The goal is to collect the following data of each FII:
##### 1 - Liquidez diária (daily liquidity)
##### 2 - Último Rendimento (last income)
##### 3 - Dividend Yield
##### 4 - Patrimônio Líquido ( Liquid Asset)
##### 5 - Rentabilidade no mês ( Profitability by Month)


```python
# Let's try to get data from 1 fii at first
soup = get_soup('abcp11')
daily_liquidity = soup.find('span',attrs={'class':'indicator-value'})
print('daily liquidity:',daily_liquidity.string)
last_income = daily_liquidity.find_next('span',attrs={'class':'indicator-value'})
print('last income:',last_income.string)
dividend_yield = last_income.find_next('span',attrs={'class':'indicator-value'})
print('dividend yield:',dividend_yield.string)
liquid_asset = dividend_yield.find_next('span',attrs={'class':'indicator-value'})
print('liquid asset:',liquid_asset.string)
proft = liquid_asset.find_next('span',attrs={'class':'indicator-value'}).find_next('span',attrs={'class':'indicator-value'})
print('profitability by month:',proft.string)
```

```python
daily_liquidity_list = []
last_income_list  = []
dividend_yield_list  = []
liquid_asset_list  = []
prof_by_month_list  = []
for i in range(len(fii_list)):
    soup = get_soup(fii_list['fii name'][i])
    daily_liquidity = soup.find('span',attrs={'class':'indicator-value'})
    last_income = daily_liquidity.find_next('span',attrs={'class':'indicator-value'})
    dividend_yield = last_income.find_next('span',attrs={'class':'indicator-value'})
    liquid_asset = dividend_yield.find_next('span',attrs={'class':'indicator-value'})
    proft = liquid_asset.find_next('span',attrs={'class':'indicator-value'}).find_next('span',attrs={'class':'indicator-value'})
    daily_liquidity_list.append(daily_liquidity)
    last_income_list.append(last_income)
    dividend_yield_list.append(dividend_yield)
    liquid_asset_list.append(liquid_asset)
    prof_by_month_list.append(proft)
    print(fii_list['fii name'][i],i)
    print(daily_liquidity.string)
    print(last_income.string)
    print(dividend_yield.string)
    print(liquid_asset.string)
    print(proft.string)
```

```python
daily_liquidity_list = []
last_income_list  = []
dividend_yield_list  = []
liquid_asset_list  = []
prof_by_month_list  = []
for i in range(len(fii_list)):
    soup = get_soup(fii_list['fii name'][i])
    daily_liquidity = soup.find('span',attrs={'class':'indicator-value'})
    last_income = daily_liquidity.find_next('span',attrs={'class':'indicator-value'})
    dividend_yield = last_income.find_next('span',attrs={'class':'indicator-value'})
    liquid_asset = dividend_yield.find_next('span',attrs={'class':'indicator-value'})
    proft = liquid_asset.find_next('span',attrs={'class':'indicator-value'}).find_next('span',attrs={'class':'indicator-value'})
    daily_liquidity_list.append(daily_liquidity.string)
    last_income_list.append(last_income.string)
    dividend_yield_list.append(dividend_yield.string)
    liquid_asset_list.append(liquid_asset.string)
    prof_by_month_list.append(proft.string)
    print('loading:'+ str(round(i/len(fii_list),2)))
```

```python
# Now let's clean the data
last_income_list
```

```python
data_lists = [
daily_liquidity_list,
last_income_list,
dividend_yield_list,
liquid_asset_list,
prof_by_month_list
]
new_dl = []
new_li = []
new_dy = []
new_la = []
new_pbm = []
new_data = [new_dl, new_li, new_dy, new_la, new_pbm]
for lista in data_lists:
    for element in lista:
        new_element = element.replace('\n','')
        new_element = new_element.strip()
        new_data[data_lists.index(lista)].append(new_element)
```
